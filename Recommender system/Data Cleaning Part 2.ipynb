```python
new_companions = []
for companions in data["companions"]:
    temp_set = set()
    lst = eval(companions)
    for companion in lst:
        if companion in remove_lst:
            continue
        elif companion in decoder:
            temp_set.update(decoder[companion])
        else:
            temp_set.add(companion)
    new_companions.append(temp_set)
```

```python
temp_min_days = []
temp_max_days = []
for row in data.itertuples():
    new_min_days = 0
    new_max_days = 0
    if row[9] in unit_converter: # If harvest unit is not in days
        new_min_days = row[7] * unit_converter[row[9]]
        new_max_days = row[8] * unit_converter[row[9]]
    else:
        new_min_days, new_max_days = row[7], row[8]
    temp_min_days.append(new_min_days)
    temp_max_days.append(new_max_days)
```

```python
data["min_harvest_days"] = temp_min_days
data["max_harvest_days"] = temp_max_days
```
